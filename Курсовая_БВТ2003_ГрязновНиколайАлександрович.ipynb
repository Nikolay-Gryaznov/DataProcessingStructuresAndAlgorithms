#### –†–µ—à–µ–Ω–∏–µ:

–ï—Å–ª–∏ n –Ω–µ—á–µ—Ç–Ω–æ - –µ—Å—Ç—å —Ä–µ—à–µ–Ω–∏–µ –±–µ–∑ –Ω–∏—á—å–∏—Ö: –∫–∞–∂–¥–∞—è –∫–æ–º–∞–Ω–¥–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–∏–≥—Ä–∞—Ç—å —Ä–æ–≤–Ω–æ n/2 –º–∞—Ç—á–µ–π –∏ –ø—Ä–æ–∏–≥—Ä–∞—Ç—å —Å—Ç–æ–ª—å–∫–æ –∂–µ. –ü—Ä–µ–¥—Å—Ç–∞–≤–∏–º —á—Ç–æ –∫–æ–º–∞–Ω–¥—ã —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω—ã –ø–æ –∫—Ä—É–≥—É, –∏ –æ–±–æ–∑–Ω–∞—á–∏–º, —á—Ç–æ i-—è –∫–æ–º–∞–Ω–¥–∞ –≤—ã–∏–≥—Ä—ã–≤–∞–µ—Ç —É n/2 —Å–ª–µ–¥—É—é—â–∏—Ö –∫–æ–º–∞–Ω–¥ –Ω–∞ –∫—Ä—É–≥–µ(–∏ –ø—Ä–æ–∏–≥—Ä—ã–≤–∞–µ—Ç –≤—Å–µ–º –æ—Å—Ç–∞–ª—å–Ω—ã–º).

–ï—Å–ª–∏ n —á–µ—Ç–Ω–æ, –Ω–∏—á—å–∏—Ö –±—ã—Ç—å –Ω–µ –º–æ–∂–µ—Ç: —Å—É–º–º–∞—Ä–Ω—ã–π —Å—á–µ—Ç –ø–æ –≤—Å–µ–º –∫–æ–º–∞–Ω–¥–∞–º –±–µ–∑ –Ω–∏—á—å–∏—Ö —Ä–∞–≤–µ–Ω 3n(n‚àí1)/2, –∏ —ç—Ç–æ —á–∏—Å–ª–æ –Ω–µ –¥–µ–ª–∏—Ç—Å—è –Ω–∞ n, –∫–æ–≥–¥–∞ n —á–µ—Ç–Ω–æ. –ö–∞–∂–¥–∞—è –Ω–∏—á—å—è —É–º–µ–Ω—å—à–∞–µ—Ç –∏—Ç–æ–≥–æ–≤—É—é —Å—É–º–º—É –Ω–∞ 1, –∏ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª-–≤–æ –Ω–∏—á—å–∏—Ö, –ø—Ä–∏ –∫–æ—Ç–æ—Ä–æ–º (3n(n‚àí1)/2)‚àít –±—É–¥–µ—Ç –¥–µ–ª–∏—Ç—å—Å—è –Ω–∞ n, —Ä–∞–≤–Ω–æ t=n/2 (—Ç–∞–∫ –∫–∞–∫ 3n(n‚àí1)/2 mod n = n/2). –ü–æ—ç—Ç–æ–º—É –µ—Å–ª–∏ –º—ã –Ω–∞–π–¥–µ–º –æ—Ç–≤–µ—Ç, –≤ –∫–æ—Ç–æ—Ä–æ–º —Ä–æ–≤–Ω–æ n/2 –Ω–∏—á—å–∏—Ö, –æ–Ω –±—É–¥–µ—Ç –æ–ø—Ç–∏–º–∞–ª—å–Ω—ã–º. –ò –µ–≥–æ –æ—á–µ–Ω—å –ª–µ–≥–∫–æ –Ω–∞–π—Ç–∏: —Å–Ω–æ–≤–∞ —Ä–∞—Å–ø–æ–ª–æ–∂–∏–º –≤—Å–µ –∫–æ–º–∞–Ω–¥—ã –Ω–∞ –∫—Ä—É–≥–µ; —Å–∫–∞–∂–µ–º, —á—Ç–æ i-—è –∫–æ–º–∞–Ω–¥–∞ –≤—ã–∏–≥—Ä—ã–≤–∞–µ—Ç —É —Å–ª–µ–¥—É—é—â–∏—Ö n‚àí2/2 –∫–æ–º–∞–Ω–¥ –≤ –∫—Ä—É–≥–µ, –ø—Ä–æ–∏–≥—Ä—ã–≤–∞–µ—Ç –ø—Ä–µ–¥—ã–¥—É—â–∏–º n‚àí2/2 –∫–æ–º–∞–Ω–¥–∞–º, –∏ –∏–≥—Ä–∞–µ—Ç –≤–Ω–∏—á—å—é —Å –ø—Ä–æ—Ç–∏–≤–æ–ø–æ–ª–æ–∂–Ω–æ–π –∫–æ–º–∞–Ω–¥–æ–π.

```python
t = int(input())
arr = []
while t:
    n = int(input())
    str = ""
    if (n % 2 == 1):
        for i in range(n):
            for j in range(i + 1, n):
                if (j - i <= n / 2):
                    str += "1 "
                else:
                    str += "-1 "
    else:
        for i in range(n):
            for j in range(i + 1, n):
                if (j - i < n / 2):
                    str += "1 "
                elif (j - i == n / 2):
                    str += "0 "
                else:
                    str += "-1 "
    t -= 1
    arr.append(str)

for i in range(len(arr)):
    print(arr[i])
```

–í —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ –æ–ø—Ç–∏–º–∞–ª—å–Ω–∞—è b —Ç–∞–∫–∞—è: 1,2,...,(k‚àím‚àí1),k,(k‚àí1),...,(k‚àím+1),(k‚àím),(k‚àím+1),...,k. –ü–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∞ p, –Ω–µ–æ–±—Ö–æ–¥–∏–º–∞—è –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è —Ç–∞–∫–æ–π b —Ä–∞–≤–Ω–∞ 1,2,...,(k‚àím‚àí1),k,(k‚àí1),...,(k‚àím).

```python
t = int(input())
arr = []
while t:
    n, k = map(int, input().split())
    a = [i for i in range(1, 2 * k - n)]
    for i in range(n - k + 1):
        a.append(k - i)
    arr.append(a)
    t -= 1

for i in range(len(arr)):
    res = ""
    for j in range(len(arr[i])):
        res += f"{arr[i][j]} "

    print(res)
```

#### –†–µ—à–µ–Ω–∏–µ

–°—Ç–æ–∏—Ç –ø–æ–¥–º–µ—Ç–∏—Ç—å –≤–∑–∞–∏–º–æ—Å–≤—è—Ç—å —Å—Ç–µ–ø–µ–Ω–∏ —á–∏—Å–ª–∞ n –¥–ª—è —Ä–µ—à–µ–Ω–∏—è –∑–∞–¥–∞—á–∏, —Ç–æ –µ—Å—Ç—å, —á—Ç–æ–±—ã –Ω–∞—Ö–æ–¥–∏—Ç—å –ø–∞—Ä—ã –Ω–∞–º —Å—Ç–æ–∏—Ç –ø–æ–º–Ω–∏—Ç—å, —á—Ç–æ 
–¥–æ–ø–æ–ª–Ω–µ–Ω–∏–µ —á–∏—Å–ª–∞ x –º–æ–∂–Ω–æ –ø–æ–ª—É—á–∏—Ç—å —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º: x‚äï(n‚àí1), –≥–¥–µ n - —ç—Ç–æ —Å—Ç–µ–ø–µ–Ω—å 2.

–¢–µ–ø–µ—Ä—å —É –Ω–∞—Å 3 —Å–ª—É—á–∞—è:

–ï—Å–ª–∏ k=0:

–í —ç—Ç–æ–º —Å–ª—É—á–∞–µ –ø–æ–ª—É—á–∞–µ–º –≤—ã—Ä–∞–∂–µ–Ω–∏–µ ‚àëx=x&c(x)=0, –≥–¥–µ x - 0‚â§x<n/2.

–ï—Å–ª–∏ 0<k<n‚àí1:

–í —ç—Ç–æ–º —Å–ª—É—á–∞–µ –º–æ–∂–Ω–æ —Å–æ–≤–º–µ—â–∞—Ç—å –∫–∞–∂–¥—ã–π —ç–ª–µ–º–µ–Ω—Ç —Å –µ–≥–æ –¥–æ–ø–æ–ª–Ω–µ–Ω–∏–µ–º, –∫—Ä–æ–º–µ 0, k, c(k) –∏ n‚àí1, –∞ –∑–∞—Ç–µ–º —Å–æ–µ–¥–∏–Ω–∏—Ç—å 
0 —Å k‚äï(n‚àí1) –ø–æ–ª—É—á–∏—Ç—Å—è 0 –∏ k —Å n‚àí1 –ø–æ–ª—É—á–∏—Ç—Å—è k.

–ï—Å–ª–∏ k=n‚àí1:

–í —ç—Ç–æ–º —Å–ª—É—á–∞–µ, –µ—Å–ª–∏ n = 4, —Ä–µ—à–µ–Ω–∏—è –Ω–µ—Ç.

–ï—Å–ª–∏ n‚â•8, –º–æ–∂–Ω–æ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –æ—Ç–≤–µ—Ç —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º:

–ú–æ–∂–Ω–æ —Å–æ–≤–º–µ—Å—Ç–∏—Ç—å n‚àí1 —Å n‚àí2, n‚àí3 —Å 1, 0 —Å 2 –∏ –≤—Å–µ –æ—Å—Ç–∞–ª—å–Ω—ã–µ —ç–ª–µ–º–µ–Ω—Ç—ã —Å –∏—Ö –¥–æ–ø–æ–ª–Ω–µ–Ω–∏—è–º–∏.

(n‚àí1)&(n‚àí2)=n‚àí2, –Ω–∞–ø—Ä–∏–º–µ—Ä 1111&1110=1110
(n‚àí3)&1=1, –Ω–∞–ø—Ä–∏–º–µ—Ä 1101&0001=0001
0 –∏ 2=0, –Ω–∞–ø—Ä–∏–º–µ—Ä 0000&0010=0000
–í—Å–µ –æ—Å—Ç–∞–ª—å–Ω—ã–µ —ç–ª–µ–º–µ–Ω—Ç—ã –º–æ–≥—É—Ç –±—ã—Ç—å —Å–æ–≤–º–µ—â–µ–Ω—ã —Å –∏—Ö –¥–æ–ø–æ–ª–Ω–µ–Ω–∏—è–º–∏ –∏ x&x‚äï(n‚àí1)=0

```python
t = int(input())
arr = []
while t:
    n, k = map(int, input().strip().split())
    if k == n - 1:
        if n == 4:
            arr.append("-1")
            continue
        arr.append(f"{n - 2} {n - 1}")
        arr.append(f"1 3")
        arr.append(f"0 {n - 4}")
        arr.append(f"2 {n - 3}")
        for x in range(4, n // 2):
            arr.append(f"{x} {n - x - 1}")
        continue
    if k == 0:
        for x in range(n // 2):
            arr.append(f"{x} {n - x - 1}")
        continue
    arr.append(f"{k} {n - 1}")
    arr.append(f"0 {n - k - 1}")
    for x in range(1, n // 2):
        if x in (k, n - k - 1):
            continue
        arr.append(f"{x} {n - x - 1}")
    t -= 1

for i in range(len(arr)):
    print(arr[i])
```

–ò–≥—Ä–∞ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç –ø–æ —Å–ª–µ–¥—É—é—â–∏–º –ø—Ä–∞–≤–∏–ª–∞–º: –≤–Ω–∞—á–∞–ª–µ –ü–µ—Ç—è —Å–ø—Ä–∞—à–∏–≤–∞–µ—Ç –≤—Å–µ –≤–æ–ø—Ä–æ—Å—ã, –∫–æ—Ç–æ—Ä—ã–µ –µ–≥–æ –∏–Ω—Ç–µ—Ä–µ—Å—É—é—Ç (–≤ —Ç–æ–º —á–∏—Å–ª–µ, –æ–Ω –º–æ–∂–µ—Ç –Ω–µ –∑–∞–¥–∞—Ç—å –Ω–∏ –æ–¥–Ω–æ–≥–æ –≤–æ–ø—Ä–æ—Å–∞), –∑–∞—Ç–µ–º –í–∞—Å—è –æ—Ç–≤–µ—á–∞–µ—Ç –Ω–∞ –∫–∞–∂–¥—ã–π –∏–∑ –≤–æ–ø—Ä–æ—Å–æ–≤ ¬´–¥–∞¬ª –∏–ª–∏ ¬´–Ω–µ—Ç¬ª. –ü–æ—Å–ª–µ –ø–æ–ª—É—á–µ–Ω–∏—è –≤—Å–µ—Ö –æ—Ç–≤–µ—Ç–æ–≤ –ü–µ—Ç—è –¥–æ–ª–∂–µ–Ω –Ω–∞–∑–≤–∞—Ç—å —á–∏—Å–ª–æ, –∫–æ—Ç–æ—Ä–æ–µ –∑–∞–≥–∞–¥–∞–ª –í–∞—Å—è.

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–í –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–ø–∏—Å–∞–Ω–æ —á–∏—Å–ª–æ n (1‚Äâ‚â§‚Äân‚Äâ‚â§‚Äâ10^3).

–ï—Å–ª–∏ –∂–µ –∑–∞–≥–∞–¥–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ –¥–µ–ª–∏—Ç—Å—è –Ω–∞ 4, —Ç–æ –æ–Ω–æ —Ä–∞–≤–Ω–æ 4.

–ï—Å–ª–∏ –∑–∞–≥–∞–¥–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ –¥–µ–ª–∏—Ç—Å—è –Ω–∞ 3, —Ç–æ –∑–∞–≥–∞–¥–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ —Ä–∞–≤–Ω–æ 3.

–ò–Ω–∞—á–µ, –æ–Ω–æ —Ä–∞–≤–Ω–æ 2. –°—Ç–∞–ª–æ –±—ã—Ç—å, —ç—Ç–∞ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –≤–æ–ø—Ä–æ—Å–æ–≤ –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ —É–≥–∞–¥—ã–≤–∞–µ—Ç –∑–∞–≥–∞–¥–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ. –ú–æ–∂–Ω–æ –ø–æ–∫–∞–∑–∞—Ç—å, —á—Ç–æ –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –≤–æ–ø—Ä–æ—Å–æ–≤ —Å–æ—Å—Ç–æ—è—â–µ–π –∏–∑ –º–µ–Ω–µ–µ, —á–µ–º —Ç—Ä—ë—Ö –≤–æ–ø—Ä–æ—Å–æ–≤, —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—è—é—â–µ–π —É—Å–ª–æ–≤–∏—é.

#### –†–µ—à–µ–Ω–∏–µ

–î–ª—è —Ä–µ—à–µ–Ω–∏—è –∑–∞–¥–∞—á–∏ –Ω—É–∂–Ω–æ —Å–ø—Ä–æ—Å–∏—Ç—å –≤—Å–µ —á–∏—Å–ª–∞ –≤–∏–¥–∞ p^k, –≥–¥–µ p ‚Äî –ø—Ä–æ—Å—Ç–æ–µ, –∞ k‚Äâ>=‚Äâ1.

```python
n = int(input())
a = [True] * (n+1)
r = []
for i in range(2, n+1):
    if a[i]:
        j = i
        while j <= n:
            a[j] = False
            j += i
        j = i
        while j <= n:
            r.append(j)
            j *= i
print(len(r))
print(*r)
```

–ï—Å–ª–∏ –Ω–∞—Ö–æ–¥–∏–º –Ω–æ–ª—å, —Ç–æ –±—É–¥–µ–º –∏–¥—Ç–∏ —Å–Ω–∞—á–∞–ª–∞ –¥–æ –∏–Ω–¥–µ–∫—Å–∞ 0 –≤–∫–ª—é—á–∞—è –µ–≥–æ - —ç—Ç–æ –±—É–¥–µ—Ç –±–æ–ª—å—à–∞—è –ø–æ–¥—Å—Ç—Ä–æ–∫–∞, –∏ –æ—Ç –Ω–∞—á–∞–ª–∞ –¥–æ –∏–Ω–¥–µ–∫—Å–∞ 0 –Ω–µ –≤–∫–ª—é—á–∞—è –µ–≥–æ - —ç—Ç–æ –±—É–¥–µ—Ç –º–µ–Ω—å—à–∞—è –ø–æ–¥—Å—Ç—Ä–æ–∫–∞ 

–ï—Å–ª–∏ —ç—Ç–æ —É—Å–ª–æ–≤–∏–µ –Ω–µ –≤—ã–ø–æ–ª–Ω–∏–ª–æ—Å—å, —Ç–æ —Å–º–µ—à–∞–µ–º –∏–Ω–¥–µ–∫—Å –≤–ª–µ–≤–æ –∏ –ø—Ä–æ–≤–µ—Ä—è–µ–º, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ –æ–Ω 1, –≤ —ç—Ç–æ–º —Å–ª—É—á–∞–µ –æ—Ç—Ä–µ–∑–æ–∫ –±–æ–ª—å—à–∏–π –ø–æ –∑–Ω–∞—á–µ–Ω–∏—é –±—É–¥–µ—Ç –æ—Ç i+1 –¥–æ n-1 –∏ –º–µ–Ω—å—à–∏–π –æ—Ç i+2 –¥–æ n (–Ω–∞–±–æ—Ä—ã –µ–¥–∏–Ω–∏—Ü, —Ç–∞–∫ –∫–∞–∫ –≤ —ç—Ç–æ–π —á–∞—Å—Ç–∏ –Ω–µ—Ç 0). –ï—Å–ª–∏ —ç–ª–µ–º–µ–Ω—Ç –Ω–µ —è–≤–ª—è–µ—Ç—Å—è 1, —Ç–æ —É –Ω–∞—Å –±—É–¥—É—Ç –ø–∞—Ä—ã –æ—Ç–ª–∏—á–∞—é—â–∏–µ—Å—è 0 –ø–µ—Ä–µ–¥ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ–º, –±–æ–ª—å—à–∏–π –æ—Ç—Ä–µ–∑–æ–∫ –æ—Ç i+1 –¥–æ n, –∞ –º–µ–Ω—å—à–∏–π i+2 –¥–æ n

```python
result = []
t = int(input())
check = False
while t:
    t -= 1
    n = int(input())
    s = input()
    for i in range(n - 1, n // 2 - 1, -1):
        if s[i] == '0':
            result.append(f"1, {i + 1}, 1, {i}")
            check = True
            break
    if check:
        check = False
        continue
    i -= 1
    if s[i] == '1':
        result.append(f"{i + 1}, {n - 1}, {i + 2}, {n}")
        continue
    else:
        result.append(f"{i + 1}, {n}, {i + 2}, {n}")
        continue

for i in range(len(result)):
    print(result[i])
```

#### –†–µ—à–µ–Ω–∏–µ

–°—Ç–æ–∏—Ç –æ–±—Ä–∞—Ç–∏—Ç—å –≤–Ω–∏–º–∞–Ω–∏–µ –Ω–∞ —Ç–æ, —á—Ç–æ –º–∞—Å—Å–∏–≤ –º–æ–∂–Ω–æ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞—Ç—å, —á—Ç–æ –≤–æ –º–Ω–æ–≥–æ–º —É–ø—Ä–æ—Å—Ç–∏—Ç —Ä–∞–±–æ—Ç—É –ø–æ –ø–æ–∏—Å–∫—É —Ç—Ä–æ–µ–∫. –î–∞–ª–µ–µ –∏–¥—ë—Ç –ø–µ—Ä–µ–±–æ—Ä —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –∏ –ø—Ä–æ—Å–º–æ—Ç—Ä –∫–æ–º–±–∏–Ω–∞—Ü–∏–π. –ù–∞—á–∞–ª–æ –∏–¥—ë—Ç —Å 0 –∏ 2 —ç–ª–µ–º–µ—Ç–Ω–æ–≤ –º–∞—Å—Å–∏–≤–∞, –ø–µ—Ä–µ–±–æ—Ä –∏–¥—ë—Ç "–≥—É—Å–µ–Ω–∏—Ü–µ–π", —Ç–æ –µ—Å—Ç—å –ø—Ä–∏ –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏–∏ –ø–æ–¥—Ö–æ–¥—è—â–µ–π(–∏—Ö) —Ç—Ä–æ–π–∫–∏(—Ç—Ä–æ–µ–∫) –ø–µ—Ä–µ—Å—Ç–∞–≤–ª—è–µ–º –ø—Ä–∞–≤—ã–π —ç–ª–µ–º–µ–Ω—Ç –Ω–∞ —Å–ª–µ–¥—É—é—â–∏–π –∏–Ω–¥–µ–∫—Å, –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ –ø–æ–¥–æ–¥–≤–∏–Ω–µ–º –ª–µ–≤—ã–π —ç–ª–µ–º–µ–Ω—Ç –Ω–∞ –∏–Ω–¥–µ–∫—Å –≤–ø—Ä–∞–≤–æ, –ø—Ä–∏ —ç—Ç–æ–º –Ω–µ –∑–∞–±—ã–≤–∞–µ–º –¥–µ–ª–∞—Ç—å –ø—Ä–æ–≤–µ—Ä–∫—É –Ω–∞ —Ä–∞–∑–Ω–∏—Ü—É –≤ —Ä–∞—Å—Ç–æ—è–Ω–∏–∏ –∏–¥–µ–∫—Å–æ–≤, –Ω–µ –º–µ–Ω–µ–µ 2-—É—Ö –ø–æ–∑–∏—Ü–∏–π.

```python
t = int(input())
arr = []
while t:
    n = int(input())
    a = sorted(list(map(int, input().split())))
    i, j, ans = 0, 2, 0
    while j < n:
        if a[j] - a[i] <= 2:
            ans += ((j - i - 1) * (j - i)) // 2
            j += 1
        else:
            i += 1
        if j - i == 1:
            j += 1
    arr.append(ans)
    t -= 1

for i in range(len(arr)):
    print(arr[i])
```

–ê–Ω—É –ø—Ä–∏–¥—É–º–∞–ª–∞ —Å–≤–æ—é —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é ùëì: ùëì(ùë•,ùë¶)=(ùë•|ùë¶)‚àíùë¶, –≥–¥–µ | –æ–±–æ–∑–Ω–∞—á–∞–µ—Ç –æ–ø–µ—Ä–∞—Ü–∏—é –ø–æ–±–∏—Ç–æ–≤–æ–≥–æ –ò–õ–ò. –ö –ø—Ä–∏–º–µ—Ä—É, ùëì(11,6)=(11|6)‚àí6=15‚àí6=9. –ú–æ–∂–Ω–æ –ø–æ–∫–∞–∑–∞—Ç—å, —á—Ç–æ –¥–ª—è –ª—é–±—ã—Ö –Ω–µ–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª ùë• –∏ ùë¶ –∑–Ω–∞—á–µ–Ω–∏–µ ùëì(ùë•,ùë¶) —Ç–∞–∫–∂–µ –Ω–µ–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–µ.

–ó–Ω–∞—á–µ–Ω–∏–µ –º–∞—Å—Å–∏–≤–∞ [ùëé1,ùëé2,‚Ä¶,ùëéùëõ] –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è –∫–∞–∫ ùëì(ùëì(‚Ä¶ùëì(ùëì(ùëé1,ùëé2),ùëé3),‚Ä¶ùëéùëõ‚àí1),ùëéùëõ) (–¥–ª—è –ø—Ä–∏–º–µ—Ä–æ–≤ –æ–±—Ä–∞—Ç–∏—Ç–µ—Å—å –∫ –ø—Ä–∏–º–µ—á–∞–Ω–∏—è–º). –í–∞–º –¥–∞–Ω –º–∞—Å—Å–∏–≤ –∏–∑ –Ω–µ –æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —á–∏—Å–µ–ª. –ö–∞–∫ –Ω—É–∂–Ω–æ –ø–µ—Ä–µ—Å—Ç–∞–≤–∏—Ç—å –º–µ—Å—Ç–∞–º–∏ –µ–≥–æ —ç–ª–µ–º–µ–Ω—Ç—ã, —á—Ç–æ–±—ã –µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏–µ —Å—Ç–∞–ª–æ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–º –≤–æ–∑–º–æ–∂–Ω—ã–º?

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\

–í—Ç–æ—Ä–∞—è —Å—Ç—Ä–æ–∫–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç ùëõ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª ùëé1,ùëé2,‚Ä¶,ùëéùëõ (0‚â§ùëéùëñ‚â§10^9)  ‚Äî —ç–ª–µ–º–µ–Ω—Ç—ã –º–∞—Å—Å–∏–≤–∞. –ù–µ –≥–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç—Å—è, —á—Ç–æ —ç–ª–µ–º–µ–Ω—Ç—ã –º–∞—Å—Å–∏–≤–∞ —Ä–∞–∑–ª–∏—á–Ω—ã.

#### –†–µ—à–µ–Ω–∏–µ

–§—É–Ω–∫—Ü–∏—é f(a,b) –º–æ–∂–Ω–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–∏—Ç—å –≤ —Å–ª–µ–¥—É—é—â–µ–º –≤–∏–¥–µ: a&(‚àºb).

–ó–Ω–∞—á–µ–Ω–∏–µ –º–∞—Å—Å–∏–≤–∞ [a1,a2,‚Ä¶,an] –±—É–¥–µ—Ç a1&(‚àºa2)&‚Ä¶(‚àºan), —á—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –ø—Ä–∏ –ª—é–±–æ–π –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –∑–∞–≤–∏—Å–∏—Ç —Ç–æ–ª—å–∫–æ –æ—Ç –ø–µ—Ä–≤–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞.

```python
n = int(input())
arr = list(map(int, input().split()))
fl = False
max_ans = 0
for i in range(n):
    ans = arr[i]
    for j in range(n):
        if i != j:
            ans = ans & (~arr[j])
            if ans <= max_ans:
                break

    if ans > max_ans:
        max_ans = ans
        idx = i
        fl = True

if fl:
    arr[0], arr[idx] = arr[idx], arr[0]
print(*arr)
```

–ù–∞–π–¥—ë—Ç—Å—è –∫—É–ø–ª–µ–Ω–Ω–æ–µ –í–∞—Å–µ–π —á–∏—Å–ª–æ l, –º–µ–Ω—å—à–µ–µ x.
–ù–∞–π–¥—ë—Ç—Å—è –∫—É–ø–ª–µ–Ω–Ω–æ–µ –í–∞—Å–µ–π —á–∏—Å–ª–æ r, –±–æ–ª—å—à–µ–µ x.

```python
import sys

t = int(input())
arr = []
k = 0
while t:
    n = int(input())
    l_value = 10 ** 10
    r_value = -1
    l_cost = 10 ** 10
    r_cost = 10 ** 10
    sum = sys.maxsize
    for i in range(n):
        l, r, c = map(int, input().split())
        if l_value > l:
            sum = sys.maxsize
            l_value = l
            l_cost = c
        elif l_value == l and l_cost > c:
            l_cost = c
        if r_value < r:
            sum = sys.maxsize
            r_value = r
            r_cost = c
        elif r_value == r and r_cost > c:
            r_cost = c
        if l_value == l and r_value == r:
            sum = min(sum, c)
        arr.append(min(l_cost + r_cost, sum))

    t -= 1

for i in range(len(arr)):
    print(arr[i])
```

–ü—Ä–µ–¥–ø–æ–ª–æ–∂–∏–º, —á—Ç–æ –Ω–∞—à –ø—É—Ç—å –±—É–¥–µ—Ç —Å–æ—Å—Ç–æ—è—Ç—å –∏–∑ —Ä–æ–≤–Ω–æ k –æ—Ç—Ä–µ–∑–∫–æ–≤. –¢–∞–∫ –∫–∞–∫ –º—ã –¥–æ–ª–∂–Ω—ã –ø—Ä–∏–π—Ç–∏ –≤ —Ç–æ—á–∫—É (n,n) –∏ –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω—ã–µ –∏ –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω—ã–µ –æ—Ç—Ä–µ–∑–∫–∏ –∏–¥—É—Ç –ø–æ –æ—á–µ—Ä–µ–¥–∏, —Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ, length1+length3+length5+‚ãØ=n –∏ legth2+length4+length6+‚ãØ=n.

–¢–∞–∫–∂–µ, –Ω–∞–º –Ω—É–∂–Ω–æ –º–∏–Ω–∏–º–∏–∑–∏—Ä–æ–≤–∞—Ç—å ‚àë—Åi=lengthi - —ç—Ç–æ –∑–Ω–∞—á–∏—Ç, —á—Ç–æ –Ω–∞–º –Ω—É–∂–Ω–æ –º–∏–Ω–∏–º–∏–∑–∏—Ä–æ–≤–∞—Ç—å c1‚ãÖlength1+c3‚ãÖlength3+‚Ä¶ –∏ c2‚ãÖlength2+c4‚ãÖlength4+‚Ä¶ –Ω–µ–∑–∞–≤–∏—Å–∏–º–æ.

–î–ª—è –ø–æ–¥—Å—á–µ—Ç–∞ –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–≥–æ –ø—É—Ç–∏, —Å–æ—Å—Ç–æ—è—â–µ–≥–æ –∏–∑ k –æ—Ç—Ä–µ–∑–∫–æ–≤, –Ω–∞–º –Ω—É–∂–Ω–æ –∑–Ω–∞—Ç—å —Å—É–º–º—É ci –Ω–∞ –Ω–µ—á–µ—Ç–Ω—ã—Ö –∏ —á–µ—Ç–Ω—ã—Ö –ø–æ–∑–∏—Ü–∏—è—Ö —Å—Ä–µ–¥–∏ c1,‚Ä¶,ck, –∞ —Ç–∞–∫–∂–µ –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–π ci –ø–æ —á–µ—Ç–Ω—ã–º –∏ –Ω–µ—á–µ—Ç–Ω—ã–º –ø–æ–∑–∏—Ü–∏—è–º. –ò–∑ —ç—Ç–æ–≥–æ —Å–ª–µ–¥—É–µ—Ç: sumOdd+minOdd‚ãÖ(n‚àícntOdd)+sumEven+minEven‚ãÖ(n‚àícntEven).

–¢–µ–ø–µ—Ä—å –Ω–∞–º –Ω—É–∂–Ω–æ –ø–µ—Ä–µ–±—Ä–∞—Ç—å –≤—Å–µ k –æ—Ç 2 –ø–æ n –∏ –Ω–∞–π—Ç–∏ –º–∏–Ω–∏–º—É–º —Å—Ä–µ–¥–∏ –æ—Ç–≤–µ—Ç–æ–≤ (–Ω–∞—Ö–æ–∂–¥–µ–Ω–∏–µ —Å—É–º–º—ã –∏ –º–∏–Ω–∏–º—É–º–æ–≤, –ø—Ä–∏ –ø–µ—Ä–µ—Ö–æ–¥–µ –æ—Ç k –∫ k+1.

```python
t = int(input())
arr = [0 for i in range(t)]
k = 0
while (t):
    n = int(input())
    min_even = 1000000
    min_odd = 1000000
    cnt_even = 0
    cnt_odd = 0
    result = 1000000
    sum = 0
    x = input().split()
    for i in range(1, n + 1):
        sum += int(x[i - 1])
        if i % 2:
            min_odd = min(min_odd, int(x[i - 1]))
            cnt_odd += 1
        else:
            min_even = min(min_even, int(x[i - 1]))
            cnt_even += 1
        result = min(result, sum + min_odd * (n - cnt_odd) + min_even * (n - cnt_even))
    arr[k] = result
    k += 1
    t -= 1

for i in range(len(arr)):
    print(arr[i])
```

–†–µ—à–µ–Ω–∏–µ —Å –ø–æ–º–æ—â—å—é –∂–∞–¥–Ω–æ–≥–æ –∞–ª–≥–æ—Ä–∏—Ç–º–∞(–ø—Ä–∏–Ω—è—Ç–∏–µ –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–≥–æ —Ä–µ—à–µ–Ω–∏—è –Ω–∞ –∫–∞–∂–¥–æ–º —ç—Ç–∞–ø–µ, –¥–æ–ø—É—Å–∫–∞–µ—Ç—Å—è –∫–æ–Ω–µ—á–Ω–æ–µ –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–µ —Ä–µ—à–µ–Ω–∏–µ). –ù–∞—á–∏–Ω–∞—è —Å 1-–≥–æ —è–π—Ü–∞ –∏ –∫–∞–∂–¥—ã–π —Ä–∞–∑ –æ—Ç–¥–∞–≤–∞–π—Ç–µ —è–π—Ü–æ A —Ç–æ–≥–¥–∞ –∏ —Ç–æ–ª—å–∫–æ —Ç–æ–≥–¥–∞, –∫–æ–≥–¥–∞ –ø–µ—Ä–µ–¥–∞—á–∞ –µ–≥–æ A –Ω–µ –∏–º–µ–µ—Ç –∑–Ω–∞—á–µ–Ω–∏—è > 500, –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ –æ—Ç–¥–∞–π—Ç–µ –µ–≥–æ G.

–£–±–µ–¥–∏—Ç—å—Å—è –≤ –ø—Ä–∞–≤–∏–ª—å–Ω–æ—Å—Ç–∏ –º–æ–∂–Ω–æ –Ω–∞ –ø—Ä–∏–º–µ—Ä–µ: –ø—Ä–µ–¥–ø–æ–ª–æ–∂–∏–º, —á—Ç–æ –º—ã –Ω–∞–∑–Ω–∞—á–∏–ª–∏ –ø–µ—Ä–≤—ã–µ n - 1 —è–∏—Ü —Ç–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, —á—Ç–æ –æ–±—â–∞—è —Å—É–º–º–∞ –¥–µ–Ω–µ–≥, –≤—ã–¥–∞–Ω–Ω—ã—Ö A, —Ä–∞–≤–Ω–∞ Sa, –∞ –æ–±—â–∞—è —Å—É–º–º–∞ –¥–µ–Ω–µ–≥, –≤—ã–¥–∞–Ω–Ω—ã—Ö G, —Ä–∞–≤–Ω–∞ Sg. –ú—ã –º–æ–∂–µ–º –ø—Ä–µ–¥–ø–æ–ª–æ–∂–∏—Ç—å, —á—Ç–æ Sa ‚â• Sg. –¢–µ–ø–µ—Ä—å –º—ã –¥–æ–ª–∂–Ω—ã –ª–∏–±–æ –¥–æ–±–∞–≤–∏—Ç—å gn –∫ Sg, –ª–∏–±–æ –¥–æ–±–∞–≤–∏—Ç—å an –∫ Sa. –ï—Å–ª–∏ –º—ã –Ω–µ –º–æ–∂–µ–º –¥–æ–±–∞–≤–∏—Ç—å gn –∫ Sg, —Ç–æ Sg + gn > Sa + 500, —Ç–∞–∫ —á—Ç–æ - 500 > Sa - Sg - gn, –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ 1000 –∫ –æ–±–µ–∏–º —Å—Ç–æ—Ä–æ–Ω–∞–º –¥–∞–µ—Ç –Ω–∞–º –Ω–µ—Ä–∞–≤–µ–Ω—Å—Ç–≤–æ 500 > Sa + (1000 - gn) - Sg, —á—Ç–æ –∏–º–µ–Ω–Ω–æ —Ç–æ, —á—Ç–æ –Ω–∞–º –Ω—É–∂–Ω–æ, —á—Ç–æ–±—ã —É–±–µ–¥–∏—Ç—å—Å—è, —á—Ç–æ –º—ã –º–æ–∂–µ–º –¥–æ–±–∞–≤—å—Ç–µ –∫ Sa = 1000 - gn.

```python
n=int(input())
a=0
g=0
str = ""
while(n>0):
    r=input().split()
    r[0]=int(r[0])
    r[1]=int(r[1])
    if(g+r[0]-a<=500):
        g+=r[0]
        str+="A"
    else:
        a+=r[1]
        str+="G"
    n-=1
print(str)
```
